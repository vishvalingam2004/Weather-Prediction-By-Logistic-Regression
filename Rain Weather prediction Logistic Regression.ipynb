```python
df["Temp9am"].fillna(df["Temp9am"].mean(skipna=True),inplace=True)
df["Temp3pm"].fillna(df["Temp3pm"].mean(skipna=True),inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
en=OrdinalEncoder()
```

```python
a=['WindGustDir']
b=['WindDir9am']
c=['WindDir3pm']
d=['RainToday']
e=['RainTomorrow']
v=['Date']
u=['Location']
```

```python
df[a]=en.fit_transform(df[a])
df[b]=en.fit_transform(df[b])
df[c]=en.fit_transform(df[c])
df[d]=en.fit_transform(df[d])
df[e]=en.fit_transform(df[e])
df[v]=en.fit_transform(df[v])
df[u]=en.fit_transform(df[u])
```

```python
df=df.astype({'RainTomorrow':'int','WindGustDir':'int','WindDir9am':'int','WindDir3pm':'int','RainToday':'int','Date':'int','Location':'int'})
```

```python
df
```

```python
cor=df.corr()
cor.style.background_gradient(cmap='Purples')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,0:25]
y=df['RainTomorrow']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scl=StandardScaler()
```

```python
xtrain=scl.fit_transform(xtrain)
xtest=scl.transform(xtest)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
mod=LogisticRegression()
mod.fit(xtrain,ytrain)
```

```python
pred=mod.predict(xtest)
print(pred)
```

```python
# Confusion Matrix.
from sklearn.metrics import confusion_matrix
```

```python
c=confusion_matrix(ytest,pred)
print("Confusion Matrix is: \n",c)
```

```python
sns.heatmap(c,annot=True,fmt='g',cmap='Blues',xticklabels=['Yes','No'],yticklabels=['Yes','No'])
plt.xlabel('Actual',fontsize=16)
plt.ylabel('Prediction',fontsize=16)
plt.title("Confusion Matrix",fontsize=20)
plt.show()
```

```python
# Evaluating Scores.
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
acc=accuracy_score(ytest,pred)
print("Accuracy is:",acc*100)
pre=precision_score(ytest,pred)
print("Precision:",pre*100)
rec=recall_score(ytest,pred)
print("Recall:",rec*100)
F1=f1_score(ytest,pred)
print("F1 Score:",F1*100)
```
